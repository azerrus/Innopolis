#### б) Выберем одну категориальную и одну числовую переменную и сформулируем гипотезу об их взаимосвязи

Определим, есть ли различия между total_day_calls для разных area_code

```python
# Построим boxplot, гистограмму и QQ-plot

fig, ax = plt.subplots(1,3, figsize = (15,4))
sns.boxplot(data = df, x = 'area_code', y = 'total_day_calls', ax = ax[0])
sns.histplot(data = df, x = 'total_day_calls', kde = True, ax = ax[1])
stats.probplot(df['total_day_calls'], dist="norm", plot = ax[2])
plt.show()
```

```python
# Тест Шапиро — Уилка
stats.shapiro(df['total_day_calls'])
```

визуально различий больших нет между категориями area_code.
по тесту Шапиро на нормальность р-значение (0,2%) < 5%.
удалим выбросы из total_day_calls: возьмем значения total_day_calls > 25

```python
# Построим boxplot, гистограмму и QQ-plot для значений total_day_calls без выбросов

fig, ax = plt.subplots(1,3, figsize = (15,4))
sns.boxplot(data = df[df['total_day_calls'] > 25], x = 'area_code', y = 'total_day_calls', ax = ax[0])
sns.histplot(data = df[df['total_day_calls'] > 25], x = 'total_day_calls', kde = True, ax = ax[1])
stats.probplot(df[df['total_day_calls'] > 25]['total_day_calls'], dist="norm", plot = ax[2])
plt.show()
```

```python
stats.shapiro(df[df['total_day_calls'] > 25]['total_day_calls'])
```

В тесте Шапиро — Уилка Н0: данные распределены нормально. 
p-value = 10,9% > уровня значимости (5%) => нельзя отвергнуть Н0 и можно сделать вывод, что данные распределены нормально

p-значение = 10,9% > 5% => данные не противоречат Н0 => total_day_calls распределен по нормальному закону.
можно использовать метод анализа дисперсии (ANOVA-тест), чтобы выявить наличие взаимосвязи количественной и качественной переменных.

```python
# ANOVA-тест для изначального датасета
area_415 = df.loc[df['area_code'] == 'area_code_415', 'total_day_calls'].values
area_408 = df.loc[df['area_code'] == 'area_code_408', 'total_day_calls'].values
area_510 = df.loc[df['area_code'] == 'area_code_510', 'total_day_calls'].values
stats.f_oneway(area_415, area_408, area_510)
```

H0 теста - среднее значение total_day_calls между группами area_code не отличается

р-значение = 9,9% > 5% => данные не противоречат Н0

Н0 не отвергается, группы area_code не отличаются по total_day_calls
