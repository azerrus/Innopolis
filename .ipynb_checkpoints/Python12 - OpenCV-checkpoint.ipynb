```python
response = urlopen("https://www.python.org/static/community_logos/python-logo.png")
image_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)
plt.imshow(image)
```
