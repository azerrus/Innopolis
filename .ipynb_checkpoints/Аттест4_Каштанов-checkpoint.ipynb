#### Выбросом считаются такие наблюдения, для которых значения переменной находятся за пределами интервала $ [Q_1 - 1.5\cdot IQR, Q_3 + 1.5\cdot IQR]$

```python
# напишем функцию, которая определяет выбросы для числовых параметров и показывает их мин и макс значения для сравнения

def outliers(df, var):
    print(f"Параметр\t\t\tинтервал выбросов\tмин / макс\t\t дельта\n")
    for i in range(len(var)):
        q1 = df[var[i]].quantile(0.25)
        q3 = df[var[i]].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5*iqr
        #if lower_bound < 0: lower_bound = 0 
        #else: lower_bound = lower_bound
        upper_bound = q3 + 1.5*iqr
        lb = lower_bound.round(1)
        ub = upper_bound.round(1)
        min = df[var[i]].min()
        max = df[var[i]].max()
        dlt_min = (min - lb).round(1)
        dlt_max = (max - ub).round(1)
        print(f"{var[i]}\t\t[{lb}, {ub}]\t\t [{min} / {max}]\t\t [{dlt_min} / {dlt_max}]")
        #df_drop = df[(df[var[i]] >= lower_bound) & (df[var[i]] <= upper_bound)]
    #return df_drop
```

```python
outliers(df_num, var = index_num)
```


По колонке дельта видно, что практически все параметры выходят за интервалы выбросов $[Q_1 - 1.5\cdot IQR,  Q_3 + 1.5\cdot IQR]$.

Используем полученные границы для параметров в дальнейшем при анализе датасета.

```python
# посчитаем количество state в каждом state_class

df_state_area.groupby('state_class')['area_sum'].count()
# количество примерно равно, поэтому можем считать, что гранифцы 90 и 100 выбраны корректно
```

```python
# создадим словарь state_class, где ключ - state, а значение - state_class

state_class = {df_state_area.index[i] + '':'' + df_state_area.iloc[i,4] for i in range(df_state_area.shape[0])}
state_class
```

```python
# в основной датасет добавим колонку state_class для дальнейшего анализа по 3 созданным группам, а не всем 51 state

df['state_class'] = df['state'].map(state_class)
df.head()
```

```python
# создадим числовую переменную для целевого показателя churn

df['churn_num'] = df['churn'].map({'yes' : 1, 'no' : 0})
df.head()
```

```python
# по отсортированным средним значениям churn по state создадим 3 группы
# создадим классы state_mean для переменной state min / med / max с границами 0 / 0.12 / 0.17

df_state_mean = pd.DataFrame(df.groupby('state')['churn_num'].mean()).reset_index()
df_state_mean['state_mean'] = np.where(df_state_mean['churn_num'] > 0.17, 'max', np.where(df_state_mean['churn_num'] > 0.12, 'med','min'))
df_state_mean['state_mean'].value_counts()
#df_state_mean.head()
```

```python
# построим барплот по средним значениям churn по state, отсортировав по убыванию

fig, ax = plt.subplots(1,1, figsize = (15,4))
sns.barplot(x = 'state', y = 'churn_num', data = pd.DataFrame(df_state_mean.sort_values(by = 'churn_num', ascending = False)), hue = 'state_mean', ax = ax)
plt.show()
```

```python
# покажем тепловую карту для переменной churn_num и новой state_mean для предварительного визуального выявления связи

plt.figure(figsize = (7,2))
sns.heatmap(pd.crosstab(df['churn_num'], df['state_mean']), annot = True, fmt = '.0f')
plt.show()
```

```python
# подготовим таблицу сопряженности для переменной churn_num и новой state_mean для выявления зависимостей

df_state_churn = pd.crosstab(df['state_mean'], df['churn'])
df_state_churn.T
```

```python
# добавим колонку с процентным отношением churn = yes к churn = no

df_state_churn = pd.merge(
    left = df_state_churn,
    right = pd.DataFrame(df_state_churn['yes'] / df_state_churn['no']).round(3),
    how = 'inner',
    left_index = True,
    right_index = True
)
df_state_churn = df_state_churn.rename(columns={0: 'perc'})
df_state_churn
```

#### По тепловой карте и дополненной таблице сопряженности видно, что наибольшую вероятность к оттоку имеют клиенты из state группы max, а наименьшую - из state группы min

```python
# создадим числовую переменную churn

df['churn_num'] = df['churn'].map({'yes' : 1, 'no' : 0})

# создадим словарь state_mean, где ключ - state, а значение - state_mean

state_mean = {df_state_mean.iloc[i,0] + '':'' + df_state_mean.iloc[i,2] for i in range(df_state_mean.shape[0])}

# добавим в датасет новую переменную, класифицирующую оттоки по state
df['state_mean'] = df['state'].map(state_mean)

# удалим предыдущую колонку state_class, которая не показала явную зависимость по оттокам
df = df.drop('state_class', axis = 1)
df.head()
```
