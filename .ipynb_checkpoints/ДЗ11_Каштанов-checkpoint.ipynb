### 2. Выберем две количественные переменные и найдем для них 95% доверительные интервалы средних значений

```python
# Оценим параметры атаки для покемонов типа Dragon и Bug

df_drag = df_new.loc[df_new['Type 1']=='Dragon', 'Attack']
df_bug = df_new.loc[df_new['Type 1']=='Bug', 'Attack']

alpha = 0.95

# Расчитаем 95% доверительный интервал для атаки Dragon
drag_mean = df_drag.mean()
drag_var = df_drag.var()
drag_num = df_drag.count()

drag_ci = stats.norm.interval(confidence = alpha, loc = drag_mean, scale = np.sqrt(drag_var/drag_num))
print(f'95% доверительный интервал для средней атаки ({drag_mean.round(1)}) покемонов типа 1 Dragon от {drag_ci[0].round(1)} до {drag_ci[1].round(1)}')

# Расчитаем 95% доверительный интервал для атаки Bug
bug_mean = df_bug.mean()
bug_var = df_bug.var()
bug_num = df_bug.count()

bug_ci = stats.norm.interval(confidence = alpha, loc = bug_mean, scale = np.sqrt(bug_var/bug_num))
print(f'95% доверительный интервал для средней атаки ({bug_mean.round(1)}) покемонов типа 1 Bug от {bug_ci[0].round(1)} до {bug_ci[1].round(1)}')

# Вывод: показатель атаки для типа Dragon выше чем для типа Bug
```

```python
# Создадим датафрейм с выбранными покемонами Dragon и Bug
df_drag_bug = df_new.set_index('Type 1', drop = True).loc[['Dragon','Bug'], :]
df_drag_bug
```

```python
# Графически оценим показатели атаки для Dragon и Bug в зависимости от Generation

sns.lineplot(data = df_drag_bug, x = 'Generation', y = 'Attack', hue = 'Type 1', errorbar = ('ci', 95))
plt.show()

# Выводы:
# Показатель атаки для типа Dragon выше чем для типа Bug.
# Для типа Dragon показатели атаки имеют больший разброс для Generation = 1, 2 и 4. Для 3, 5, 6 - меньший.
# Для типа Bug показатели атаки имеют больший разброс для Generation = 1 и 2. Для 3, 4, 5, 6 - меньший.
# И для Dragon, и для Bug средние показатели атаки наименьшие для Generation = 6.
```

### 3. Оценим распределение выбранных переменных на нормальность с помощью формальных тестов и с помощью альтернативных методов

```python
fig = plt.figure()
plt.title('Распределение атаки покемонтов типа Dragon')
sns.histplot(df_drag, kde = True, binwidth = 20)
plt.vlines(x = drag_ci[0], ymin = 0, ymax = 9, color = 'r')
plt.vlines(x = drag_ci[1], ymin = 0, ymax = 9, color = 'r')
plt.vlines(x = drag_mean, ymin = 0, ymax = 9, color = 'g')
plt.show()
```

```python
print('Средняя: ', df_drag.mean())
print('Медиана: ', df_drag.median())
print('Скошенность: ', stats.skew(df_drag))
print('Эксцесс: ', stats.kurtosis(df_drag))
```

Скошенность и эксцесс близки к нулю => распределение можно считать нормальным

```python
fig = plt.figure()
plt.title('Распределение атаки покемонтов типа Bug')
sns.histplot(df_bug, kde = True, binwidth = 25)
plt.vlines(x = bug_ci[0], ymin = 0, ymax = 18, color = 'r')
plt.vlines(x = bug_ci[1], ymin = 0, ymax = 18, color = 'r')
plt.vlines(x = bug_mean, ymin = 0, ymax = 18, color = 'g')
plt.show()
```

```python
print('Средняя: ', df_bug.mean())
print('Медиана: ', df_bug.median())
print('Скошенность: ', stats.skew(df_bug))
print('Эксцесс: ', stats.kurtosis(df_bug))
```

Скошенность >> 0 => распределение нельзя считать нормальным

##### Тест Шапиро-Уилка

```python
stats.shapiro(df_drag)
stats.shapiro(df_bug)
print(f'p-значение по тесту Шапиро_Уилка для Dragon = {stats.shapiro(df_drag)[1].round(2)}')
print(f'p-значение по тесту Шапиро_Уилка для Bug = {stats.shapiro(df_bug)[1].round(4)}')
```

для Dragon p-значение = 0,61 (>0,098), т.е. нулевая гипотеза принимается и распределение атаки МОЖНО считать нормальным

для Bug p-значение = 0,0091 (<0,098), т.е. нулевая гипотеза отвергается и распределение атаки НЕЛЬЗЯ считать нормальным

##### Тест Андерсона-Дарлинга

```python
stats.anderson(df_drag, dist='norm')
print(f'Для Dragon по тесту Андерсона-Дарлинга статистика = {stats.anderson(df_drag, dist='norm')[0].round(3)}, критическое значение для 5% = {list(stats.anderson(df_drag, dist='norm')[1].round(3))[2]}')
stats.anderson(df_bug, dist='norm')
print(f'Для Bug по тесту Андерсона-Дарлинга статистика = {stats.anderson(df_bug, dist='norm')[0].round(3)}, критическое значение для 5% = {list(stats.anderson(df_bug, dist='norm')[1].round(3))[2]}')
```

##### Тест Колмогорова-Смирнова

```python
stats.kstest(df_drag, stats.norm.cdf)
stats.kstest(df_bug, stats.norm.cdf)
print(f'Для Dragon по тесту Колмогорова-Смирнова р-значение = {stats.kstest(df_drag, stats.norm.cdf)[1].round(2)}')
print(f'Для Bug по тесту Колмогорова-Смирнова р-значение = {stats.kstest(df_bug, stats.norm.cdf)[1].round(2)}')
```

Для Dragon р-значение является очень низким, т.е. нулевая гипотеза отвергается. Распределение атаки НЕ является нормальным

Для Bug р-значение является очень низким, т.е. нулевая гипотеза отвергается. Распределение атаки НЕ является нормальным

Вывод: по большинству тестов распределение атаки для Dragon МОЖНО считать нормальным, для Bug - НЕЛЬЗЯ

### 4. Возьмем две группы наблюдений и протестируем гипотезу о равенстве математических ожиданий (двухвыборочный t-тест или тест Манна-Уитни в зависимости от результата теста на нормальность)

#### Тест о равенстве мат ожиданий (двухвыборочный t-тест)

Протестируем гипотезу, что средняя атака покемонов типа Dragon и Bug совпадают

```python
stats.ttest_ind(df_drag, df_bug, alternative = 'two-sided')
print(f'p-значение = {stats.ttest_ind(df_drag, df_bug, alternative = 'two-sided')[1].round(3)}')
```

p-значение оказалось ниже 5%, т.е. нулевая гипотеза о том, что средние значения атаки для Dragon и Bug равны, ОТВЕРГАЕТСЯ

#### Тест Манна-Уитни

```python
stats.mannwhitneyu(df_drag, df_bug, alternative = 'two-sided')
print(f'р-значение = {stats.mannwhitneyu(df_drag, df_bug, alternative = 'two-sided')[1].round(3)}')
```

р-значение существенно меньше 5% => нулевая гипотеза теста о том, что формы распределения атак для Dragon и Bug совпадают (в т.ч., что совпадают их медианы), отвергается. Медианные значения атаки не равны!

### 6. Оценим модель линейной регрессии. Прогноз и вывод по результатам

```python
# Оценим линейную регрессию для статов HP и Attack для покемонов типа 1 Dragon и типов 2 = Flying, Ground, Phychic
X = np.array(df_fly_ground_phys['HP'])
Y = np.array(df_fly_ground_phys['Attack'])
X = sm.add_constant(X)
model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

Для статов HP и Attack для покемонов тип 1 Dragon и типов 2 = Flying, Ground, Phychic выявили линейную регрессию:
Yi = 9.43 + 1.26 * Xi
Выводы: 
1. найдена линейная взаимосвязь между показателями HP и Attack. Коэффициент 𝛽_1 - положительный => при увеличении значений HP растут показатели Attack
2. глядя на точечную диаграмму 'HP - Attack' линейная зависимость хорошо описывается при диапазонах значений HP [50;65] и [85;105], при значениях HP <50, [65;80] и >105 имеется значительный разброс значений Attack.
