```python
months = {
    'January' : '01',
    'February' : '02',
    'March' : '03', 
    'April' : '04',
    'May' : '05',
    'June' : '06',
    'July' : '07',    
    'August' : '08',
    'September' : '09',
    'October' : '10',
    'November' : '11', 
    'December' : '12'
}
```

```python
long_df['Month_num'] = long_df['Month'].map(months)
```

```python
long_df['Year_Month'] = long_df['Year'].astype(str) +'-' + long_df['Month_num']
long_df
```

```python
def combine_strings(str1, str2):
    return str1 + '-' + str2
```

```python
combine_strings('aaa', 'bbb')
```

```python
long_df
```

```python
long_df.apply(lambda x: combine_strings(str(x['Year']), x['Month_num']), axis = 1)
```
