### 2. Напишим другую функцию и применим ее к сгруппированным данным с помощью .transform()


```python
def get_min(series):
    min = series.min()
    return min

def get_max(series):
    max = series.max()
    return max
```

```python
df_long['Stat_min'] = df_long.groupby('Stats')['Amount'].transform(get_min)
df_long['Stat_max'] = df_long.groupby('Stats')['Amount'].transform(get_max)
df_long.drop_duplicates(subset = 'Stats')
```

### 4. Сгруппируем данные по дате (т.к. неприменимо к нашему датасету используем выборку из БД demo) и используем resample для агрегирования в более крупный периода


```python
import psycopg2
```

```python
query = """
WITH flights_per_day AS
(
	WITH russian_airports_table AS
	(
		SELECT
			fl.departure_airport,
			fl.arrival_airport,
			fl.actual_departure,
			t_fl.amount
		FROM 
			flights AS fl
		LEFT JOIN
			ticket_flights AS t_fl
		ON t_fl.flight_id = fl.flight_id
		WHERE fl.actual_departure IS NOT NULL AND t_fl.amount IS NOT NULL
		GROUP BY
			fl.actual_departure,
			fl.departure_airport,
			fl.arrival_airport,
			t_fl.amount
	)
	SELECT
		DATE(actual_departure),
		actual_departure,
		to_char(actual_departure, 'ID'::text)::integer AS days_of_week,
		SUM (CASE
				WHEN departure_airport IN ('SVO', 'VKO', 'DME') OR arrival_airport IN ('SVO', 'VKO', 'DME') THEN 1 ELSE 0 END) AS MOW,
		SUM (CASE
				WHEN departure_airport NOT IN ('SVO', 'VKO', 'DME') OR arrival_airport NOT IN ('SVO', 'VKO', 'DME') THEN 1 ELSE 0 END) AS non_MOW,
		amount
	FROM
		russian_airports_table
	GROUP BY
		DATE(actual_departure),
		actual_departure,
		days_of_week,
		amount
)
SELECT
	DATE(actual_departure),
	days_of_week,
	mow,
	non_mow,
	mow + non_mow AS total,
	amount,
	AVG (amount) OVER (PARTITION BY days_of_week) AS avg_amount_per_day
FROM
	flights_per_day
GROUP BY
	DATE(actual_departure),
	actual_departure,
	mow,
	non_mow,
	days_of_week,
	amount
ORDER BY
	actual_departure"""
```

```python
engine = psycopg2.connect("dbname='demo' user='postgres' host='localhost' port='5432' password='12kshtanov'")
```

```python
df_demo = pd.read_sql(query, con=engine)
```
