### 1. Выберем на сайте http://www.lib.ru любой крупный текст

```python
# Выберем произведение "Приглашение на охоту" Джордж Хичкок.
# Проведем парсинг текста в строку

url = "http://www.lib.ru/INOFANT/HICHKOK/hitchcock_an_invitation_to_the_hunt.txt_with-big-pictures.html"
response = requests.get(url)
soup = bs4.BeautifulSoup(response.text, 'html')
text = soup.body.get_text(' ', strip = True)
text
```

```python
# Подготовим текст:
# сделаем строчные буквы, уберем знаки препинания и числа
text.lower().translate(str.maketrans('', '', string.punctuation)).translate(str.maketrans('', '', string.digits))
```

```python
# Разделим строку текста на отдельные слова
words = text.lower().translate(str.maketrans('', '', string.punctuation)).translate(str.maketrans('', '', string.digits)).split()
```

```python
# Проведем лемматизацию слов
lemma = [morph.parse(word)[0].normal_form for word in words]
lemma
```

### 2. Определим 10 самых часто встречающихся слов

```python
# Найдем топ-10 самых частовстречающихся слов по тексту

keyword = db.from_sequence(lemma)
keyword_count = keyword.frequencies()
word_count = pd.DataFrame(keyword_count, columns=['Word', 'Frequency']).sort_values(by="Frequency", ascending = False).head(10).reset_index(drop = True)
word_count
```

#### Видно, что из текста необходимо выбросить неключевые слова

### 2.1 Определим 10 самых часто встречающихся СУЩЕСТВИТЕЛЬНЫХ

```python
# Через атрибут POS определим часть речи.
# В первом случае выберем из текста только существительные.

lemma_noun = [morph.parse(word)[0].normal_form for word in words if morph.parse(word)[0].tag.POS == 'NOUN']
lemma_noun
```

```python
# Найдем топ-10 самых часто встречающихся слов по тексту

keyword_noun = db.from_sequence(lemma_noun)
keyword_count = keyword_noun.frequencies()
word_count_noun = pd.DataFrame(keyword_count, columns=['Word', 'Frequency']).sort_values(by="Frequency", ascending = False).head(10).reset_index(drop = True)
word_count_noun
```

Среди только существительных часто встречаются имена главных героев

### 2.2 Определим 10 самых часто встречающихся ОСНОВНЫХ ЧАСТЕЙ РЕЧИ

```python
# Во втором случае выберем только слова, не являющиеся предлогами, местоимениями, союзами, частицами и междометиями

lemma_main = [morph.parse(word)[0].normal_form for word in words if (morph.parse(word)[0].tag.POS != 'PREP')&(morph.parse(word)[0].tag.POS != 'NPRO')&(morph.parse(word)[0].tag.POS != 'CONJ')&(morph.parse(word)[0].tag.POS != 'PRCL')&(morph.parse(word)[0].tag.POS != 'INTJ')]
lemma_main
```

```python
# Найдем топ-10 самых часто встречающихся слов по тексту

keyword_main = db.from_sequence(lemma_main)
keyword_count_main = keyword_main.frequencies()
word_count_main = pd.DataFrame(keyword_count_main, columns=['Word', 'Frequency']).sort_values(by="Frequency", ascending = False).head(10).reset_index(drop = True)
word_count_main
```

### 2.3 Определим 10 самых часто встречающихся ДЛИННЫХ СЛОВ

```python
# В третьем случае выберем слова длиной более 3 символов

lemma_long = [word for word in lemma if len(word) > 3]
lemma_long
```

### 3. Построим гистограмму распределения


```python
# Гистограмма распределения для любых слов (изначальный текст)

plt.figure(figsize = (15,4))
plt.bar(word_count.loc[:,'Word'].values, height = word_count.loc[:,'Frequency'].values)
plt.title('Частота упоминания любых слов в тексте')
for i in range(len(word_count)):
    plt.annotate(text = f'({word_count.iloc[i,1]})', xy = (word_count.loc[:,'Word'].values[i], word_count.loc[:,'Frequency'].values[i]))

plt.show()
```

```python
# Гистограмма распределения только для существительных

plt.figure(figsize = (15,4))
plt.bar(word_count_noun.loc[:,'Word'].values, height = word_count_noun.loc[:,'Frequency'].values)
plt.title('Частота упоминания только существительных в тексте')
for i in range(len(word_count_noun)):
    plt.annotate(text = f'({word_count_noun.iloc[i,1]})', xy = (word_count_noun.loc[:,'Word'].values[i], word_count_noun.loc[:,'Frequency'].values[i]))
plt.show()
```

```python
# Гистограмма распределения для основных частей речи

plt.figure(figsize = (15,4))
plt.bar(word_count_main.loc[:,'Word'].values, height = word_count_main.loc[:,'Frequency'].values)
plt.title('Частота упоминания основных частей речи в тексте')
for i in range(len(word_count_main)):
    plt.annotate(text = f'({word_count_main.iloc[i,1]})', xy = (word_count_main.loc[:,'Word'].values[i], word_count_main.loc[:,'Frequency'].values[i]))
plt.show()
```

```python
# Гистограмма распределения для любых слов длиннее 3 символов

plt.figure(figsize = (15,4))
plt.bar(word_count_long.loc[:,'Word'].values, height = word_count_long.loc[:,'Frequency'].values)
plt.title('Частота упоминания длинных слов в тексте (> 3 символов)')
for i in range(len(word_count_long)):
    plt.annotate(text = f'({word_count_long.iloc[i,1]})', xy = (word_count_long.loc[:,'Word'].values[i], word_count_long.loc[:,'Frequency'].values[i]))
plt.show()
```

```python
# Построим 4 предыдущие гистограммы рядом для сравнения

fig, ax = plt.subplots(1,4, figsize = (15,4))
ax[0].bar(word_count.loc[:,'Word'].values, height = word_count.loc[:,'Frequency'].values)
ax[1].bar(word_count_noun.loc[:,'Word'].values, height = word_count_noun.loc[:,'Frequency'].values)
ax[2].bar(word_count_main.loc[:,'Word'].values, height = word_count_main.loc[:,'Frequency'].values)
ax[3].bar(word_count_long.loc[:,'Word'].values, height = word_count_long.loc[:,'Frequency'].values)
plt.suptitle('Распределение слов для разных категорий', fontsize = 15)
ax[0].set_title('Распределение по всем словам', fontsize = 8)
ax[1].set_title('Распределение по существительным', fontsize = 8)
ax[2].set_title('Распределение по основным частям речи', fontsize = 8)
ax[3].set_title('Распределение по длинным словам', fontsize = 8)
for ax in fig.axes:
    plt.sca(ax)
    plt.xticks(rotation=90)
plt.show()
```
