Выбросом считаются такие наблюдения, для которых значения переменной находятся за пределами интервала $ [Q_1 - 1.5IQR, Q_3 + 1.5IQR]$

```python
# найдем выбросы для параметров
def outliers(df, var):
    print(f"Параметр\t\t\tинтервал выбросов\tмин / макс")
    for i in range(len(var)):
        q1 = df[var[i]].quantile(0.25)
        q3 = df[var[i]].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5*iqr
        #if lower_bound < 0: lower_bound = 0 
        #else: lower_bound = lower_bound
        upper_bound = q3 + 1.5*iqr
        print(f"{var[i]}\t\t[{lower_bound.round(1)}, {upper_bound.round(1)}]\t\t [{df[var[i]].min()} / {df[var[i]].max()}]")
        #df_drop = df[(df[var[i]] >= lower_bound) & (df[var[i]] <= upper_bound)]
    #return df_drop
```

Видно, что некоторые параметры выходят за интервалы выбросов $ [Q_1 - 1.5IQR,  Q_3 + 1.5IQR]$.

#### 2) Используем дополнительные переменные в модели (без нормализации)

```python
cat_vars = ['international_plan', 'voice_mail_plan']
num_vars = ['number_vmail_messages', 'number_customer_service_calls']
target_var = ['target']
```

```python
# Переводим категориальные переменные в числовой формат
X_dummies = pd.get_dummies(df_new[cat_vars], drop_first = True, dtype = int)
X_dummies.head()
```

```python
# Объединяем таблицу
X = df_new[num_vars]
X = pd.merge(
    left = X,
    right = X_dummies,
    how = 'inner',
    left_index = True,
    right_index = True
)

y = df_new['target']
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42)
X_test, X_forecast, y_test, y_forecast = train_test_split(X_test, y_test, test_size = 0.5, random_state = 42)
```

```python
Prec_mean = []
Rec_mean = []
for i in range(3, 13, 1):
    neigh = KNeighborsClassifier(n_neighbors = i)
    neigh.fit(X_train, np.ravel(y_train))
    y_pred = neigh.predict(X_test)
    print('Neigbors:', i, 
          'Precision:', np.round(precision_score(y_test, y_pred),3), 
          'Recall:', np.round(recall_score(y_test, y_pred),3)
         )
    Prec_mean.append(precision_score(y_test, y_pred))
    Rec_mean.append(recall_score(y_test, y_pred))
print('Precision_mean:', np.round(np.mean(Prec_mean),3),'Recall_mean:', np.round(np.mean(Rec_mean),3))
```

При добавлении категориальных переменных international_plan и voice_mail_plan без нормализации параметры модели изменились

Precision упал, Recall вырос
