```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
ax[0].pie(sizes, labels = labels, autopct = '%0.2f%%', startangle = 180)
ax[0].set_title('Круговая')
ax[1].bar(labels, sizes)
ax[1].set_title('Столбчатая')
plt.suptitle('Комбинированный график')
plt.tight_layout()
plt.show()
```
