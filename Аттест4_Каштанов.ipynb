#### Выбросом считаются такие наблюдения, для которых значения переменной находятся за пределами интервала $ [Q_1 - 1.5\cdot IQR, Q_3 + 1.5\cdot IQR]$

```python
# напишем функцию, которая определяет выбросы для числовых параметров и показывает их мин и макс значения для сравнения

def outliers(df, var):
    print(f"Параметр\t\t\tинтервал выбросов\tмин / макс\t\t дельта\n")
    for i in range(len(var)):
        q1 = df[var[i]].quantile(0.25)
        q3 = df[var[i]].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5*iqr
        #if lower_bound < 0: lower_bound = 0 
        #else: lower_bound = lower_bound
        upper_bound = q3 + 1.5*iqr
        lb = lower_bound.round(1)
        ub = upper_bound.round(1)
        min = df[var[i]].min()
        max = df[var[i]].max()
        dlt_min = (min - lb).round(1)
        dlt_max = (max - ub).round(1)
        print(f"{var[i]}\t\t[{lb}, {ub}]\t\t [{min} / {max}]\t\t [{dlt_min} / {dlt_max}]")
        #df_drop = df[(df[var[i]] >= lower_bound) & (df[var[i]] <= upper_bound)]
    #return df_drop
```

```python
outliers(df_num, var = index_num)
```


По колонке дельта видно, что практически все параметры выходят за интервалы выбросов $[Q_1 - 1.5\cdot IQR,  Q_3 + 1.5\cdot IQR]$.

Используем полученные границы для параметров в дальнейшем при анализе датасета.
