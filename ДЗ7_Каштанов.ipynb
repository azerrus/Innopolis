### 5. Проверить, есть ли в датасете выбросы

```python
df_new.describe()
```

Выбросом считаются такие наблюдения, для которых значения переменной находятся за пределами интервала $ [Q_1 - 1.5IQR, Q_3 + 1.5IQR]$

```python
# найдем выбросы для параметра жизнь 'HP'
def drop_outliers(df, variable = 'HP'):
    q1 = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    print(f"Выбросами считаются значения параметра 'HP', которые находятся за пределами интервала [{lower_bound}, {upper_bound}]")
    return df[(df['HP']>=lower_bound)&(df['HP']<=upper_bound)]
```

```python
df_no_outliers_HP = drop_outliers(df_new)
df_no_outliers_HP
```

```python
df_new['HP'].hist(bins=20)
```

```python
df_no_outliers_HP['HP'].hist(bins=15)
```

### 7. Нормализовать данные при помощи способа "Нормализация"


Приведение числовой величины к такому масштабу, чтобы минимальное значение соответствовало 0, а максимальное - единице (1)

$$ \tilde{x}_i = \frac{x_i - min}{max - min} $$

```python
pd.options.mode.copy_on_write = True
```

```python
df_no_outliers_HP['norm_HP'] = (df_no_outliers_HP['HP'] - df_no_outliers_HP['HP'].min()) / (df_no_outliers_HP['HP'].max() - df_no_outliers_HP['HP'].min())
df_no_outliers_HP
```

```python
df_no_outliers_HP['norm_HP'].describe()
```

```python
df_no_outliers_HP['norm_HP'].hist(bins = 15)
```
